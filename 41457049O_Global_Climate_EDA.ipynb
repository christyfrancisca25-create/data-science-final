```python
The objective of this project is to analyze global temperature anomalies since 1880 and identify a 10-year rolling trend.
```

Initialize: Import pandas for data manipulation, matplotlib for plotting, and numpy for data generation.

Data Generation: Create a sequence of years from 1880 to 2025 and generate a synthetic temperature anomaly with a warming trend and random noise.

Process Algorithm: Apply a 10-year rolling average to the anomaly data to smooth the noise and highlight the long-term trend.

Visualization: Create a scatter plot of the annual data and overlay the 10-year trend line to visualize the change over time.

Summary: Print the most recent anomaly value to summarize the current state of warming.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. DATA LOADING

# Creating a sample dataset so you can see the results immediately
years = np.arange(1880, 2025)
temp_anomaly = 0.008 * (years - 1880) + np.random.normal(0, 0.1, len(years)) - 0.4
df = pd.DataFrame({'Year': years, 'Anomaly': temp_anomaly})

# 2. ALGORITHM: 10-YEAR ROLLING AVERAGE---This algorithm to smooth out the noise and show the climate trend
df['Rolling_Avg'] = df['Anomaly'].rolling(window=10).mean()

# 3. VISUALIZATION 
plt.figure(figsize=(12, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# Plot raw data and the smoothed trend
plt.scatter(df['Year'], df['Anomaly'], color='lightgray', label='Annual Anomaly', s=10)
plt.plot(df['Year'], df['Rolling_Avg'], color='red', linewidth=2, label='10-Year Trend')

# Formatting the chart
plt.title('Global Temperature Anomaly (1880-2025)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Deviation from Baseline (C)', fontsize=12)
plt.axhline(0, color='black', linestyle='--', alpha=0.5) # Baseline
plt.legend()

plt.show()

# 4. OUTPUT SUMMARY
latest_anomaly = df['Anomaly'].iloc[-1]
print(f"Current Temperature Anomaly: {latest_anomaly:.2f}C")
```
